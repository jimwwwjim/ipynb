```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#for 10 intervals
mu=2
sigma=1.1
t=1.5
n1=10
X0=1
dt=t/n1

rn=np.random.standard_normal(n1)
w=np.cumsum(rn)*np.sqrt(dt)

X=X0+mu*dt+sigma*w
T=np.linspace(dt,t,n1)

plt.plot(T,X)
```
